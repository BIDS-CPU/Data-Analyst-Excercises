## **1.How to identify the Records (Order ID + Product ID combination) present in data1 but missing in data2**

```python
Q1=df1[df1['primary_key'].isin(df2['primary_key'])==False]
Q1
```

```python
print('Number_of_missing_records',len(Q1))
```

## **2.How to identify the Records (Order ID + Product ID combination) missing in data1 but present in data2**

```python
Q2=df2[df2['primary_key'].isin(df1['primary_key'])==False]
Q2
```

```python
print('Number_of_missing_records',len(Q2))
```

# **3.Find the Sum of the total Qty of Records missing in data1 but present in data2**


```python
Q3=Q2['Qty'].sum()
print('total Qty of Records missing in data1 but present in data2',Q3)
```

## **4.Find the total number of unique records**

```python
Q4=df1.merge(df2,indicator=True)
Q4
```

```python
print('Find the total number of unique records',len(Q4))
```
